```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)
   
    return mplleaflet.display()


    
    
    
leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
tmp = 'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89'
df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/{}.csv'.format(tmp))
```

```python
# create dataframe with 2005-2014 data
df_one = df[pd.DatetimeIndex(df['Date']).year != 2015]
```

```python
# create dataframe with 2015 data
df_two = df[pd.DatetimeIndex(df['Date']).year == 2015]
```

```python
df_min = df_one.groupby(['Day', 'Element'])['Data_Value'].min()
df_max = df_one.groupby(['Day', 'Element'])['Data_Value'].max()
```

```python
df_two_min = df_two.groupby(['Day', 'Element'])['Data_Value'].min()
df_two_max = df_two.groupby(['Day', 'Element'])['Data_Value'].max()
```

```python
lista = list(range(1,366))
data2015 = pd.DataFrame(lista, index = lista, columns=['TMIN'])
```

```python
for i in range(1,366):
    data2015.loc[i]['TMIN'] = df_two_min[i]['TMIN']
```

```python
data2015 = data2015.assign(TMAX = lista)
```

```python
for i in range(1,366):
    data2015.loc[i]['TMAX'] = df_two_max[i]['TMAX']
```

```python
#df_minimum = pd.DataFrame
#for i in range(1,366):
#    df_minimum = 
lista = list(range(1,366))
prova = pd.DataFrame(lista, index = lista, columns=['TMIN'])
```

```python
for i in range(1,366):
    prova.loc[i]['TMIN'] = df_min[i]['TMIN']
```

```python
prova = prova.assign(TMAX = lista)
```

```python
for i in range(1,366):
    prova.loc[i]['TMAX'] = df_max[i]['TMAX']
```

```python
record_low = data2015[data2015['TMIN']<prova['TMIN']]['TMIN']
record_high = data2015[data2015['TMAX']>prova['TMAX']]['TMAX']
```
